```python
import matplotlib.pyplot as plt
import math as m
x = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
plt.figure(figsize=(12,6))

# y = 0n + 1 and label the curve "O1"
y1 = [0*n + 1 for n in x]
plt.plot(x,y1, label='O1')

# y = log(n) and lebel the curve 'O(log n)'
y2 = [m.log(n) for n in x]
plt.plot(x,y2, label='O($log n$)')

# y = n and label the curve "O(n)"
y3 = [n for n in x]
plt.plot(x,y3, label='O($n$)')

# y = n*log(n) and label it "O(n log n)"
y4 = [n*m.log(n) for n in x]
plt.plot(x,y4, label='O($n log n$)')

# y = n**2 and label it "O(n^2)"
y5 = [n**2 for n in x]
plt.plot(x,y5, label='O($n^2$)')

# y = 2**n and label it "O(2^n)"
y6 = [2**n for n in x]
plt.plot(x,y6, label='O($2^n$)')

# y = n! and label it "O(n!)"
y7 = [m.factorial(n) for n in x]
plt.plot(x,y7, label='O($n!$)')

# y = n**n and label it "O(n^n)"
y8 = [n**n for n in x]
plt.plot(x,y8, label='O($n^n$)')

plt.xlim(1, 8)
plt.ylim(0, 100)

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
```
