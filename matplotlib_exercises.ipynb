##### 1. Use matplotlib to plot the following equation: y=x**2 - x + 2
##### You'll need to write the code that generates the x and y points.

##### Add an anotation for the point 0, 0, the origin.

```python
x = list(range(0,10))
y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.annotate('origin', xy=(0,5), xytext = (0, 20), arrowprops={'facecolor':'orchid'})
plt.show()
```

```python
y1 = [n**.5 for n in x]
plt.plot(x,y1, label='sqrt(x)')

y2 = [n**3 for n in x]
plt.plot(x,y2, label = 'x^3')

x3 = range(0,10)
y3 = [2**n for n in x]
plt.plot(x,y3, label = '2^x')

y4 = [1/(n+1) for n in x]
plt.plot(x,y4, label = '1/(x+1)')

plt.title('Plotted Equations')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
